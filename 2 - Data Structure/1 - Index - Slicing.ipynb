Slicing

```python
barcode = "ABCDE12541"
```

```python
# barcode[starting index : stopping index : stepping size]

# Starting Index → Specifies which index to start from.
# Stopping Index → Specifies at which index to stop.
# Stepping Size → Specifies how many steps to move each time.
```

```python
barcode[3::]
```

```python
barcode[:3:]
```

```python
barcode[::3]
```

```python
barcode[3:5]
```

```python
barcode[1:9:2]
```

```python
# If you set the step size to -1, it reverses the data.
barcode[::-1]
```
