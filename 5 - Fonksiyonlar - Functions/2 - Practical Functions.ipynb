```python
def divide_number(number):
    return number / 2
```

```python
divide_number(20)
```

```python
my_list = [3, 5, 7, 10, 20, 30]
```

```python
my_result_list = []
```

```python
for num in my_list:
    my_result_list.append(divide_number(num))
```

```python
my_result_list
```

```python
list(map(divide_number, my_list))
```
