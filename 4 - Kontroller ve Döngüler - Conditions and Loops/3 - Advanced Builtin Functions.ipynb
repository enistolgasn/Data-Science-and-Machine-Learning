```python
list(range(5, 20, 2))
```

```python
my_list = [20, 30, 40, 50, 60]
```

```python
# When we use a normal for loop, the elements of the list are printed.
for num in my_list:
    print(num)
```

```python
# If we want to print the indices, we can use the range function.
for ix in range(len(my_list)):
    print(ix)
```

ENUMERATE

```python
for element in enumerate(my_list):
    print(element)
```

```python
for (ix, value) in enumerate(my_list):
    print(ix, value)
```

ZIP

```python
food_list = ["apple", "banana", "melon"]
```

```python
calories_list = [100, 150, 200]
```

```python
day_list = ["monday", "tuesday", "wednesday"]
```

```python
zipped_list = zip(food_list, calories_list, day_list)
```

```python
zipped_list
```

```python
type(zipped_list)
```

```python
zipped_list = list(zip(food_list, calories_list, day_list))
```

```python
zipped_list
```
