```python
# With the insert method, we can specify where the value will be added. The code below will add the value 90 at index 2.
my_list.insert(2, 90)

my_list
```

```python
# The pop method removes the last value from the list.
my_list.pop()

my_list
```

```python
# The remove method removes the specified value from the list.
my_list.remove(90)

my_list
```

```python
# The reverse method reverses the list.
my_list.reverse()
```
