```python
class Musician():
    def __init__(self, name):
        self.name = name
        print("musician class")

    def test1(self):
        print("test1")

    def test2(self):
        print("test2")
```

```python
musician = Musician("John")
```

```python
musician.name
```

```python
musician.test1()
```

```python
musician.test2()
```

```python
class MusicianPlus(Musician):
    def __init__(self, name):
        super().__init__(name)
        print("musician plus")
```

```python
musician_plus = MusicianPlus("Clark")
```

```python
musician_plus.test1()
```

```python
musician_plus.test2()
```
