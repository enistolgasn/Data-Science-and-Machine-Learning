```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
    @abstractmethod
    def car_name(self):
        pass
```

```python
class Tesla(Car):
    def car_name(self):
        print("Tesla")
```

```python
tesla = Tesla()
```

```python
tesla.car_name()
```

```python
class Mercedes(Car):
    def car_name(self):
        print("Mercedes")
```

```python
mercedes = Mercedes()
```

```python
mercedes.car_name()
```
