```python
class Banana():
    def __init__(self, name):
        self.name = name

    def info(self):
        return f"100 calories {self.name}"
    
class Apple():
    def __init__(self, name):
        self.name = name

    def info(self):
        return f"150 calories {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruit_list = [banana, apple]
```

```python
for fruit in fruit_list:
    print(fruit.info())
```
