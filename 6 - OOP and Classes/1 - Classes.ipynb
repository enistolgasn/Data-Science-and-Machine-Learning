```python
class Person():
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def print_name(self):
        print(self.name)
```

```python
person = Person("John", 70, "Male")
```

```python
person.name
```

```python
person.age
```

```python
person.gender
```

```python
person.print_name()
```
