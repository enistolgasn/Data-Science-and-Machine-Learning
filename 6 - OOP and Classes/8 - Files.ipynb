```python
my_file = open("myfile.txt")
```

```python
type(my_file)
```

```python
# The read method advances the cursor step by step and outputs every character in the file until it reaches the end.
my_file.read()
```

```python
# Since we just used read and moved the cursor all the way to the end of the file, when we call the read method again, the cursor is already at the end of the file, so it returns an empty value because there is nothing left to read.
my_file.read()
```

```python
# We can return to the beginning of the file using seek(0).
my_file.seek(0)
```

```python
# To avoid dealing with such manual operations, with is used.

with open("myfile.txt") as my_file:
    my_content = my_file.read()
```

```python
my_content
```

```python
# w â†’ allows us to write to the file with write permission.
with open("myfile.txt", mode="w") as my_new_file:
    my_new_file.write("test4")
```

```python
my_new_file
```

```python
with open("myfile.txt", mode="r") as my_file2:
    my_content = my_file2.read()
```

```python
my_content
```

```python
# If we want to append to a file, we use the a mode.
with open("myfile.txt", mode="a") as my_new_file3:
    my_new_file3.write("test5")
```
