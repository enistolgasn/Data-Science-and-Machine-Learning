```python
class Phone():
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def info(self):
        print(f"{self.name} price is: {self.price}")
```

```python
iphone = Phone("iPhone", 500)
```

```python
iphone.info()
```

```python
iphone.price = 400
```

```python
iphone.info()
```

```python
class Phone2():
    def __init__(self, name2, price2):
        self.name2 = name2
        # By adding two underscores before price, we ensure that it can only be accessed from within the Phone class.
        self.__price2 = price2

    def info(self):
        print(f"{self.name2} price is: {self.__price2}")
```

```python
iphone2 = Phone2("iPhone", 500)
```

```python
iphone2.info()
```

```python
iphone2.price = 400
```

```python
iphone2.info()
```
