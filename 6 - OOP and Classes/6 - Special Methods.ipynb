```python
class Fruit():
    def __init__(self, name, calories):
        self.name = name
        self.calories = calories

    def __str__(self):
        return f"{self.name}: {self.calories} calories"
```

```python
my_fruit = Fruit("banana", 150)
```

```python
my_fruit.calories
```
